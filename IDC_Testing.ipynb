```python
from cdapython import Q, columns, unique_terms, query
import pprint
```

<font color=blue>IDC Demo Query 1 from Elijah<font>

```python
disease = Q('ResearchSubject.Specimen.primary_disease_type = "Cystic, Mucinous and Serous Neoplasms"')
gdc = Q('ResearchSubject.identifier.system = "GDC"')
idc = Q('identifier.system = "IDC"')
pdc = Q('ResearchSubject.identifier.system = "PDC"')

q = idc.From(pdc.From(disease.And(gdc)))
r = q.run()

r
```

<font color=blue> IDC Demo query 2</font>

```python
diease = Q('ResearchSubject.associated_project = "TCGA-BRCA"')
age = Q('days_to_birth < -50*365')
idc = Q('identifier.system = "IDC"')

q = idc.From(disease.And(age))
r = q.run()

print(r)
```

<font color=blue>IDC Demo query 3</font>

```python
subject = Q('id = "TCGA-13-1409"')
r = subject.run()
print(r)
```

```python
count = 0
for case in r:
    filelist = case['File']
    for file in filelist:
        if file['data_category'] == 'Imaging':
            count += 1
print(str(count))
```

```python
r.pretty_print(0)
```

```python
subject = Q('id = "TCGA-13-1409"')
filetype = Q('File.data_category = "Imaging"')

q = subject.And(filetype)
r = q.run()

print(r)
```

```python
format = Q('File.file_format = "DICOM"')
q = subject.And(filetype.And(format))
r = q.run()
print(r)
```

```python
q = filetype.From(subject)
r = q.run()
print(r)
```
